# Outlier Detection
## Detect the data points that deviate significantly from the majority of the data

```python
# Z score, Percentile, IQR - removal method

sns.boxplot(df.bmi);
```

```python
q1 = df.bmi.quantile(0.25)
q3 = df.bmi.quantile(0.75)
```

```python
q1, q3
```

```python
upperlimit = q3 + 1.5 * (q3 - q1)
lowerlimit = q1 - 1.5 * (q3-q1)

upperlimit, lowerlimit
```

```python
# Now remove all the data with the bmi value greater than 'upperlimit'

condition = df['bmi'] > upperlimit
condition
```

```python
filtered_df = df[~condition]
sns.boxplot(filtered_df.bmi)
```

```python
p99 = df.bmi.quantile(0.99)
p99
```

```python
filtered_data = df['bmi'] < p99
filtered_data
filtered_df = df[filtered_data]
filtered_df
```

```python
sns.boxplot(filtered_df.bmi)
```

```python
# Z score

from scipy import stats

bmi_zscore = stats.zscore(df.bmi)
bmi_zscore
```

```python
df_z = df['bmi'] < (np.abs(bmi_zscore) <= 3)
df_z
```

```python
sns.boxplot(df_z.bmi)
```
