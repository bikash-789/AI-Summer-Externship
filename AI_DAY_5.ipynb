```python
# Let's create a data frame
import pandas as pd
import numpy as np

# Here, columns are the keys and rows are the values of each key
data = {
    "Name" : ['Bikash', 'Ayush', 'Bhuwan'],
    "Gender" : ['M', 'M', 'M'],
    "Org" : ['VIT', 'SRM', 'JAIN'],
    "Age": [np.nan, 22, 21]
}

df = pd.DataFrame(data)
df
```

```python
# One way of defining Dataframe
df = {
    'Apples' : [35, 21],
    'Bananas' : [41, 34],
}
df = pd.DataFrame(df, index=['2017 Sales', '2018 Sales'])
df
```

```python
# Another way of defining Dataframe
df = {
    'Apple' : [35, 21],
    'Banana' : [41, 34],
    'Sales' : ['2017 Sales', '2018 Sales']
}
df = pd.DataFrame(df) # first creating column named Sales which has index for data
df.set_index('Sales', inplace=True) # second setting index by col name
df
```

```python
# Now let us see how indexing is done in dataframe with the help of pandas
# iloc - based on index value selection which means what is the numerical location of any row or col
df.iloc[0]
```

```python
# The second braces selects the column based on index value. Like 0 - 1st col, 1 - 2nd col, 3 - 3rd col
df.iloc[0][0]
```

```python
# Let's create a dataframe with more number of rows and cols

df = {
    'Name': ['John', 'Emma', 'Michael', 'Sophia', 'David'],
    'Age': [25, 30, 35, 28, 32],
    'Country': ['USA', 'Canada', 'UK', 'Australia', 'Germany'],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'ID': [101, 102, 103, 104, 105]
}

df = pd.DataFrame(df)
df
```

```python
# Let us set index as the ID col
df.set_index('ID', inplace=True)
df
```

```python
# Now let's try out index based selection

df.iloc[0]  # will select all the col values of first row
```

```python
df.iloc[:3, :3] # will select first 3-rows and first 3-cols
```

```python
df.iloc[:, -1:]   # will select all the rows with last col values
```

```python
df.iloc[:, -3:]
```

```python
df
```

```python
# Let us see the another way of selection which is label based selection
df.loc[102, :] # will select all the cols value of the row with index = 102
```

```python
# will select the row with index value = 102 and col with names 'Salary', 'Name', 'Country'
df.loc[102, ['Salary', 'Name', 'Country']] 
```

```python
# will select all the rows with index values in the list and the cols values
df.loc[[101, 103, 105], :]
```

```python
df
```

```python
# Conditional Selection

# will produce a Series of True/False boolean based on the Salary value
df.Salary >= 60000
```

```python
# Now let us select those rows which satisfy the above condition i.e. Salary >= 60000
df.loc[df.Salary >= 60000]
```

```python
# Let us use combined logic to select records
# will select the record having age in between 25 and 30 and has a Salary >= 50000
df.loc[(df.Salary >= 50000) & (df.Age >= 25) & (df.Age <= 30)] # we can combine as many logic using '&' operator
```
