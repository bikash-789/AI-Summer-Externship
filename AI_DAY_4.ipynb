```python
# Series and Dataframe in pandas

# Series - single column
# Dataframe - multiple column


import pandas as pd


a = np.arange(1, 20, 2)
a

s = pd.Series(a)
s
```

```python
# Indexing in Series
s[4]
```

```python
# In Series, we can change the index values
s.index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
s
```
