```python
mat = [[1, 2, 3, 4], [4, 5, 6, 7], [6, 7, 8, 9]]

for i in mat:
  for j in i:
    print(f"{j} ", end="")
  print("\n")
```

```python
class Person:
  def __init__(self, name, age, gender):
    self.name = name
    self.age = age
    self.gender = gender
  
  def greet(self):
    print(f"Hello I'm {self.name}. I'm {self.age} years old")
  

person1 = Person("Bikash", 22, "Male");
person1.greet()
```
